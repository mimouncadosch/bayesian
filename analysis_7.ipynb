## Updating posteriors in 2D (multivariate case) 

```python
## %matplotlib inline
import scipy.stats as stats
from IPython.core.pylabtools import figsize
from IPython.core.interactiveshell import InteractiveShell
import numpy as np
import matplotlib.pyplot as plt
figsize(12.5, 4)
# InteractiveShell.ast_node_interactivity = "all"
%matplotlib inline
```

### Setup

We have $[\; y_1, y_2\;]$, a bivariate normally distributed population.

$$
\begin{bmatrix} y_1\\ y_2 \end{bmatrix} 
\sim 
\mathcal (  \begin{bmatrix} \theta_1\\ \theta_2 \end{bmatrix}, 
            \begin{bmatrix} 1 & \rho \\ \rho & 1 \end{bmatrix} )
$$

### Prior distribution (uniform)

$$
\begin{bmatrix} \theta_1\\ \theta_2 \end{bmatrix}
\sim 
Uniform(p)
$$

```python
# Range of prior values
n_points = 100
theta_1s = theta_2s = np.linspace(0, 5, n_points)
# Prior disributions
theta_1 = stats.uniform.pdf(theta_1s, loc=0, scale=5)
theta_2 = stats.uniform.pdf(theta_2s, loc=0, scale=5)
```

```python
# Bivariate prior distribution P(theta_1, theta_2)
prior = np.dot(theta_1[:, None], theta_2[None, :])
```

```python
plt.imshow(prior, origin='lower', extent=(0, 5, 0, 5));
```

### Data observed

```python
# Number of samples observed
N = 10

theta_1_true = 1
theta_2_true = 3

sigma11 = 0.3
sigma22 = 0.4
sigma12 = 0.2

cov = [[sigma11**2, sigma12**2], 
       [sigma12**2, sigma22**2]]

data = stats.multivariate_normal.rvs(mean=[theta_1_true, theta_2_true], cov=cov, size=(N, 1))
```

### Likelihood (normally distributed data)

```python
# # For a single point
# stats.multivariate_normal.pdf(data, mean=[0.5, 2.8], cov=cov)
# stats.multivariate_normal.pdf(data, mean=[1, 2.5], cov=cov)
```

Observe that the likelihood for the means $\theta_1$, $\theta_2$ close to our data is highest.  
Could also find this graphically.

```python
likelihood = np.zeros((n_points, n_points))
total_prob = 0
for id1, t1 in enumerate(theta_1s):
    for id2, t2 in enumerate(theta_2s):
        likelihood[id1, id2] = stats.multivariate_normal.pdf(data, mean=[t1, t2], cov=cov).prod()
        total_prob += likelihood[id1, id2] * prior[id1, id2]
```

```python
plt.xlabel(r'$\theta_1$')
plt.ylabel(r'$\theta_2$')
plt.imshow(likelihood, origin='lower', cmap=plt.cm.jet, extent=(0, 5, 0, 5));
```

```python
unnormalized_posterior = prior * likelihood
```

```python
posterior = prior * likelihood / total_prob
```

```python
plt.xlabel(r'$\theta_1$')
plt.ylabel(r'$\theta_2$')
plt.imshow(posterior, cmap=plt.cm.jet, origin='lower', extent=(0, 5, 0, 5));
```
