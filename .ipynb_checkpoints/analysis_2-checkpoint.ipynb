```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pymc as pm
```

### Read data

```python
# Read data
# Mine Seam  9-38
# Depth (meters) 40-46
# Height (meters)  48-54
# Width (meters)  56-62
# width/height ratio  64-70 
# Roadway width  (meters)  72-78
# Uniaxial compression strength  (MPa)  80-86
# Strength  (MPa)  88-94
# Stress (MPa)    96-102
# Strength/Stress ratio   104-110
# Stability   118   /* 1=Yes  0=No  */
# http://www.stat.ufl.edu/~winner/data/pillar_stability.txt
data = np.genfromtxt("data/pillar_stability.txt", usecols=[0, 1, 2, 6, 8, 9])
```

```python
data.shape
```

```python
data = pd.DataFrame(data, columns=['depth', 'height', 'width', 'strength', 'young_mod', 'stability'])
```

```python
plt.scatter(data.strength, data.stability)
plt.show()
```

We want to model the probability of the pillar being stable as a function of its strength.

We use a logistic regression to model this probability:
$$p(s) = \frac{1}{ 1 + e^{ \;\beta s + \alpha} }$$
where $s$ is the strength

### Playing with variables of logistic regression

```python
def logistic(x, beta, alpha):
    return 1.0 / (1.0 + np.exp(np.dot(beta, x) + alpha))

x = np.linspace(-4, 4, 100)
plt.plot(x, logistic(x, 1, 0), label="$\beta = 1$")
plt.plot(x, logistic(x, -1, 0), label="$\beta = 1$")
plt.plot(x, logistic(x, -10, 0), label="$\beta = 1$")
plt.show()
```

```python
plt.plot(x, logistic(x, -0.75, 5), label="$\beta = 1$", ls="--")
plt.plot(x, logistic(x, -10, 10), label="$\beta = 1$", ls="--")
plt.plot(x, logistic(x, -10, 20), label="$\beta = 1$", ls="--")
plt.show()
```

### Modeling our distribution

The probability distribution of stability as a function of stregnth is given by the logistic regression.
The logistic regression is parametrized by $\alpha$ and $\beta$. We can model these parameters with random variables. 

Model parameters $\alpha$ and $\beta$ with normal random variables, since they have no reason to be positive, bounded, or relatively large.

```python
beta = pm.Normal("beta", 0, 0.001, value=0)
alpha = pm.Normal("alpha", 0, 0.001, value=0)

@pm.deterministic
def p(t=data.strength, alpha=alpha, beta=beta):
    return 1.0 / ( 1. + np.exp(beta * t + alpha) )
```

Pillar stability is modeled with a Bernouilli random variable, where the probability $p$ of stability is given by the logistic regression, in turn a function of pillar strength.
$$
D_i \sim \text{Ber}( \;p(s_i)\; ), \;\; i=1..N
$$

```python
observed = pm.Bernoulli("bernouilli_obs", p, data.stability, observed=True)
```

```python
model = pm.Model([observed, beta, alpha])
```

```python
map_ = pm.MAP(model)
map_.fit()
```

```python
mcmc = pm.MCMC(model)
mcmc.sample(120000, 100000, 2)
```

### Posterior distributions
Show the posterior distributions for parameters $\beta$ and $\alpha$.

```python
# Histogram of posterior distributions
alpha_samples = mcmc.trace("alpha")[:, None]
beta_samples = mcmc.trace("beta")[:, None]

ax1 = plt.subplot(211)
sns.distplot(beta_samples, ax=ax1, label=r"posterior of $\beta$", color="blue", kde=False)
ax1.legend(loc="upper left")

ax2 = plt.subplot(212)
sns.distplot(alpha_samples, ax=ax2, label=r"posterior of $\alpha$", color="green", kde=False)
ax2.legend(loc="upper left")

plt.show()
```

### Understanding result of MCMC
The Monte Carlo Markov Chain returns random samples from the posterior distributions of $\alpha$ and $\beta$. These random samples are contained in the `alpha_samples` and `beta_samples` arrays.

### Plotting the results
__For every strength s of interest, sample beta at random, sample alpha at random, from their respective posterior distributions, and compute probability of stability based on beta and alpha sampled, for that stability value s.__

```python
s = np.linspace(data.strength.min() - 5, data.strength.max() + 5, 50)[:, None]
p_s = logistic(s.T, beta_samples, alpha_samples)
mean_prob_s = p_s.mean(axis=0)
```

```python
plt.scatter(data.strength, data.stability)
plt.plot(s, mean_prob_s, lw=3, label="average posterior probability of stability")
plt.plot(s, p_s[0], ls="--", label="realization from posterior")
plt.plot(s, p_s[2], ls="--", label="realization from posterior")

plt.title("Posterior expected value of probability of stability, plus realizations")
plt.show()
```

There are 10,000 realizations, one for every possible value in the posterior distributions of $\beta$ and $\alpha$.

```python
from scipy.stats.mstats import mquantiles

qs = mquantiles(p_s, [0.025, 0.975], axis=0)
plt.fill_between(s[:, 0], *qs, alpha=0.7, color="#7A68A6")

plt.plot(s, mean_prob_s, lw=1, ls="--", label="average posterior probability of stability")

plt.plot(s[:, 0], qs[1], label="2.5% CI", color="#fa8000", alpha=0.7)
plt.plot(s[:, 0], qs[0], label="95% CI", color="#7A68A6", alpha=0.7)
plt.xlabel('Strength s')
plt.ylabel('Probability of stability $p(s)$')
plt.scatter(data.strength, data.stability)
plt.show()
```

Notice that the credible interval is tighter (thus more confident) for extreme strengths, so our model is credible for either very low or very high strengh values. The credible interval is wider for the values of strength hovering between 5 and 10, meaning that it is less credible / confident around these values whether the pillar is stable or not. 

```python
# Compute the probability of stability for stability s = 13, 
# for all betas and alphas in posterior distributions
prob_13 = logistic(13, beta_samples, alpha_samples)

sns.distplot(prob_13, kde=False)
plt.title('Probability of stability for s = 13')
plt.show()
```

```python
# Compute the probability of stability for stability s = 13, 
# for all betas and alphas in posterior distributions
prob_5 = logistic(5, beta_samples, alpha_samples)

sns.distplot(prob_5, kde=False)
plt.title('Probability of stability for s = 5')
plt.show()
```

For s = 5, much more uncertainty about stability, in accordance to posterior expected values of probablity of stability.
